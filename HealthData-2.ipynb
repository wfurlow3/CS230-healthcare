```python
import pandas as pd

encounters = pd.read_csv("data/encounters.csv")
encounters["ENCOUNTER"] = encounters["Id"]
observations = pd.read_csv("data/observations.csv")
conditions = pd.read_csv("data/conditions.csv")

merged = conditions.merge(encounters, on = ["ENCOUNTER", "PATIENT"], how = "inner", suffixes = ["_con", "_enc"])
final = merged.merge(observations, on = ["ENCOUNTER", "PATIENT"], how = "inner")
# columns that vary per row within an encounter
varying_cols = ['CATEGORY', 'CODE', 'DESCRIPTION', 'VALUE', 'UNITS', 'TYPE']

# columns that remain constant within an encounter
constant_cols = [c for c in final.columns if c not in varying_cols]

# group by encounter
json_data = (
    final.groupby("ENCOUNTER").apply(
        lambda g: {
            **g[constant_cols].iloc[0].drop(labels="ENCOUNTER").to_dict(),  # shared info
            "MEASUREMENTS": g[varying_cols].to_dict(orient="records")        # nested list
        }
    ).to_dict()
)

#Three example patients to view
json_data["1516d2e6-4846-5f1e-fe27-c1ebb9a39f72"]
#json_data["8ccaea8a-8401-6e54-03f7-c754e7a236bb"]
#json_data["33782d5d-ba30-c2e5-c432-e36a8f6c1cc8"]
```
