```python
from typing import Optional
import re

def clean_fragment(value: Optional[object]) -> Optional[str]:
    if value is None or (isinstance(value, float) and np.isnan(value)):
        return None
    text = str(value).strip()
    if not text:
        return None
    text = re.sub(r"[^A-Za-z0-9]+", "_", text.upper())
    text = text.strip("_")
    return text or None
```

```python
unique_cond_codes = set(clean_fragment(c) for c in conditions["CODE"] if not pd.isna(c))
print("Unique condition codes in raw conditions.csv:", len(unique_cond_codes))
```
