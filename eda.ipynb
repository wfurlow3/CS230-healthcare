```python
import pandas as pd
```

```python
encounters = pd.read_csv('data/encounters.csv')
```

```python
len(encounters)
```

```python
print(encounters.columns)
```

```python
for col in encounters.columns:
    null_pct = encounters[col].isnull().mean() * 100
    print(f"{col}: {null_pct:.2f}% null")
```

```python
encounters.head()
```
